```python
#initialize transform
transform = transforms.Compose([
    transforms.Resize((428,321)),
    transforms.ToTensor()
])
```

## Data Example

This section establishes the data loader and prints an example training image.

```python
#initialize dataloader
root_dir = 'data/card'
batch_size = 45
ygoDset = dataset.ygoCards(root_dir=root_dir,transform = transform)
ygoLoader = DataLoader(ygoDset, batch_size=batch_size, shuffle=True, num_workers=4)
```

```python
#preview data
batch = next(iter(ygoLoader))
trans = transforms.ToPILImage()
plt.imshow(trans(batch[0]))
```

## Network Parameters

Control various parameters of this network. We recommend looking at the DCGAN architecture or looking at the provided Yu-GAN-Oh! code to look closer at the Generator and Discriminator class definitions.

### latent_size
Length of random Noise input to generator.

### num_gan_features
Controls width of generator and discriminator.
A higher number corresponds to a larger amount of convolutional kernels.

### num_hidden features
Number of output features for the Discriminator. 

### lr
Learning rate. We found .0001 to be a fair number.

### beta1
Used for gradient momentum calculation. Check hyperparameters of Adam optimizer in PyTorch documentation for more details.

### num_epochs
Number of epochs. 100 is typically enough to be near convergence, however we did not try training past 200 epochs due to compute time.

### similarity_features
Number of similarity features for learned similarity matrix to be used for minibatch discrimination. 

### lower_bound and bound
Lower bound of value to be used for label smoothing. This sets true labels to be between lower_bound and 1 at random.

```python
#network params
latent_size = 100
num_gan_features = 64
num_disc_features = num_gan_features
num_hidden_features = 256
lr = .0001
beta1 = 0.5
num_epochs = 100
similarity_features = 50
lower_bound = 0.8
bound = 1- lower_bound
```

## Initialize Network and Loss

Initialize network and loss parameters. Print network and confirm that network is on the device.

```python
#initialize network
device = torch.device('cuda:0')
gen = yuGANoh_with_fc_and_disc.Generator(latent_size,num_gan_features).to(device)
disc = yuGANoh_with_fc_and_disc.Discriminator(num_disc_features,num_hidden_features,similarity_features,batch_size).to(device)
gen.apply(yuGANoh_with_fc_and_disc.init_weights)
print(gen)
disc.apply(yuGANoh_with_fc_and_disc.init_weights)
print(disc)
```

```python
#initialize loss and optimizer
criterion = nn.BCELoss()
noise = torch.randn(1,latent_size,1,1).to(device)
optimizerD = optim.Adam(disc.parameters(),lr=lr,betas=(beta1,0.999), weight_decay = .0001)
optimizerG = optim.Adam(gen.parameters(),lr=2*lr,betas=(beta1,0.999), weight_decay = .0001)
```

## Feature Loss

We include the option to train using feature loss. This uses the full feature tensors to train instead of cross-entropy on the labels. This allows the training process to be more stable as feature vectors become more uniform.

```python
def feature_loss(real_features,fake_features):
    temp = torch.mean(real_features,axis=0) - torch.mean(fake_features,axis=0)
    return torch.sum(temp*temp)
```

## Debug/Util Functions

These functions are used to debug the training loop or transform the data in some manner.
We keep check_norm as a metric to check that the gradient is not vanishing during the training process.

```python
def check_norm(model):
    total_norm = 0
    for p in model.parameters():
        if p.grad is None:
            continue
        param_norm = p.grad.data.norm(2)
        total_norm += param_norm.item() ** 2
    total_norm = total_norm ** (1. / 2)
    print(total_norm)
    
def normalize_tensors(inp):
    return (inp-0.5)*2

def undo_normalize(inp):
    return inp/2 + 0.5

def check_individual_norm(model):
    for p in model.parameters():
        if p.grad is None:
            print(str(p.name)+' is None')
            continue
        print(str(p.name) +': '+  str(p.grad.data.norm(2).item()**2))
```

## Training loop

Standard DCGAN training loop with some optimizations.

```python
images = []
gen_losses = []
disc_losses = []

print("Begin training")
for epoch in range(num_epochs):
    for i,img in enumerate(ygoLoader,0):
        #zero discriminator gradient
        disc.zero_grad()
        
        
        #run real image through discriminator
        img = img.to(device)
        #img = normalize_tensors(img) + 0.1*torch.randn_like(img).to(device)
        if epoch < 15:
            img = img + 0.1*torch.randn_like(img).to(device)
        b_size = img.shape[0]
        features_real, output = disc(img)
        output = output.squeeze()
        #real_label = 1
        #label = torch.ones_like(output).to(device)
        label = 1 - bound*torch.rand_like(output)
        disc_loss_real = criterion(output,label)
        disc_loss_real.backward(retain_graph = True)
        
        
        #now run a fake batch through generator.
        inp = torch.randn(b_size,latent_size,1,1).to(device)
        gen_out = gen(inp)
        features_fake, fake_out = disc(gen_out)
        fake_out = fake_out.squeeze()
        #fake_label = 0
        label_fake = label*0
        disc_loss_fake = criterion(fake_out,label_fake)
        disc_loss_fake.backward(retain_graph = True)
        total_loss = disc_loss_real + disc_loss_fake       
        optimizerD.step()
        
        #train the generator
        #we already have a generator pass with gen_out=gen(inp)
        gen.zero_grad()
        model_loss = feature_loss(features_real,features_fake)
        model_loss.backward(retain_graph = True)   
        optimizerG.step()
        '''
        #train the generator again
        #we already have a generator pass with gen_out=gen(inp)
        gen.zero_grad()
        features_out_real , model_out = disc(gen_out)
        model_out = model_out.squeeze()
        model_loss = feature_loss(features_out_real,features_fake)
        model_loss.backward()   
        optimizerG.step()'''
        
        #check_individual_norm(gen)
        #check_individual_norm(disc)
        
        if i%50 == 0:
            disc_losses.append(total_loss.item())
            gen_losses.append(model_loss.item())
            print('Epoch: '+str(epoch) + ' iter: ' + str(i) + ' lossG: ' + str(gen_losses[-1]) + ' lossD: ' + str(disc_losses[-1]))
            out_img = gen(noise)
            #out_img = undo_normalize(out_img)
            trans = transforms.ToPILImage()
            plt.imshow(trans(out_img[0].cpu()))
            if gen_losses[-1] == 0.0:
                break
            if disc_losses[-1] == 0.0:
                break
            print('discriminator gradient L2 Norm')
            check_norm(disc)
            print('generator gradient L2 Norm')
            check_norm(gen)
```
